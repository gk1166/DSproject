```python
# 1. Write a program which will find factors of given number and 
# find whether the factor is even or odd.
# Hint: Use Loop with if-else statements
def print_factors(x):
    print("The factors of",x,"are:")
    for i in range(1, x + 1):
        if x % i == 0:
            if (i % 2 ==0):
                print(i,"is even")
            else:
                print(i,"is odd")
num =int(input("enter the any number: "))
print_factors(num)
```

```python
# 3. Write a program, which will find all the numbers between 1000 and 3000 
# (both included) such that each digit of a number is an even number. 
# The numbers obtained should be printed in a comma separated sequence 
# on a single line.
# Hint: In case of input data being supplied to the question, 
#     it should be assumed to be a console input. 
#     Divide each digit with 2 and verify is it even or not.
l=int(input("enter the starting range"))
u=int(input("enter the ending range"))

odd=" "
for m in range(l,u+1):
    for i in str(m):
        if(int(i) % 2 != 0):
            odd="true"
    if(odd!= "true"):
        print(m)
    odd=" "

```

```python
#4. Write a program that accepts a sentence and calculate the number of letters and digits.
# Suppose if the entered string is: Python0325
# Then the output will be:
# LETTERS: 6
# DIGITS:4
a,d=0,0
input_str=input("enter any word with numbers: ")
for i  in input_str:
    if(i.isdigit()):
        d=d+1
    elif(i.isalpha()):
        a=a+1
    else:
        pass
print("letters: ",a)
print("digits: ",d)
```
