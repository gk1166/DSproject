```python
# 1. You are given a dataset, which is present in the LMS, containing the number of hurricanes occurring in the United States along the coast of the Atlantic. Load the data from the dataset into your program and plot a Bar Graph of the data, taking the 
# Year as the x-axis and the number of hurricanes occurring as the Y-axis.
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_csv("Hurricanes.csv")
x=df['Year']
y=df['Hurricanes']
fig=plt.figure(figsize=(10,5))
plt.bar(x,y,color='maroon')
plt.xlabel("YEARS")
plt.ylabel('HURRICANES')
plt.title('BAR PLOT')
plt.show()
```

```python
# 2. The dataset given, records data of city temperatures over the years’ 2014 
# and 2015. Plot the histogram of the temperatures over this
# period for the cities of San Francisco and Moscow.
df=pd.read_csv('CityTemps.csv')
sanfrancisco=df['San Francisco']
moscow=df['Moscow']
plt.hist([sanfrancisco,moscow],5,rwidth=0.95,color=['green','orange'])
plt.xlabel('temperature----orange-moscow , green-sanfrancisco')
plt.ylabel('frequency')
plt.title("temperature in cities")
plt.show()
```
